```python
import streamlit as st
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pickle

# Load Data
@st.cache_data
def load_data():
    df = pd.read_csv(r"C:\Users\afshe\Desktop\stroke\stroke prediction dataset.csv")  # Ensure correct path
    return df

df = load_data()

# Streamlit App Layout
st.title("Stroke Prediction Data Analysis")

# Introduction
st.header("Introduction")
st.write("This project analyzes the Stroke Prediction Dataset to identify key insights and build a predictive model.")

# EDA Section
st.header("Exploratory Data Analysis (EDA)")
st.write("### Dataset Overview")
st.dataframe(df.head())

st.write("### Missing Data Analysis")
missing_percentage = df.isnull().mean() * 100
fig, ax = plt.subplots()
missing_percentage.sort_values(ascending=False).plot(kind='bar', color='salmon', ax=ax)
plt.title("Percentage of Missing Data by Column")
st.pyplot(fig)

st.write("### Gender Distribution")
gender_counts = df['gender'].value_counts()
fig, ax = plt.subplots()
gender_counts.plot(kind='bar', color=['skyblue', 'lightgreen', 'salmon'], ax=ax)
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Distribution of Gender in Dataset')
st.pyplot(fig)

# Model Section
st.header("Machine Learning Model")
model = None
try:
    with open("stroke_model.pkl", "rb") as file:
        model = pickle.load(file)
except FileNotFoundError:
    st.write("Model file not found. Ensure stroke_model.pkl is in the directory.")

if model:
    st.write("### Predict Stroke Risk")
    age = st.number_input("Enter Age:", min_value=0, max_value=100, value=50)
    avg_glucose = st.number_input("Enter Average Glucose Level:", min_value=0.0, value=100.0)
    bmi = st.number_input("Enter BMI:", min_value=0.0, value=25.0)
    
    if st.button("Predict"):
        prediction = model.predict([[age, avg_glucose, bmi]])
        st.write(f"Predicted Stroke Risk: {'Yes' if prediction[0] == 1 else 'No'}")

# Conclusion
st.header("Conclusion")
st.write("This project provides insights into stroke prediction using data analysis and machine learning.")
```
